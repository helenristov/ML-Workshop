```
from pyspark.sql.functions import *

#Define workshop fs path
ws_path = "/mnt/channelsa-datasets-DONOTEDIT/End-to-End-ML-Lifecycle/"

#Create a few fake columns

#FOR BASIC

#Column that shows when the device was last serviced
df_train = df_train.withColumn("serviceDate", expr("concat(cast (rand()*(12) as int) + 1, cast (rand()*(3) as int) + 2016 )"))

#Column that shows when the most recent incident happened
df_train = df_train.withColumn("recentIncident", expr("cast (rand()*(94694398) as int) + 1451606401"))

#Rename 2 columns in the basic training dataset
df_train = df_train.withColumnRenamed("Platform", "platfrm").withColumnRenamed("Processor", "prcsr")

#FOR ENRICH

#Add . markers in the Census_PrimaryDiskTotalCapacity column
#Previously, something that looks like 999999, now looks like 999.999.0
#Something that looked like 1999999, now looks like 1999.999.0

df_train = df_train.withColumn("Census_PrimaryDiskTotalCapacity", reverse("Census_PrimaryDiskTotalCapacity")).withColumn("Census_PrimaryDiskTotalCapacity", concat(lit("0"), col("Census_PrimaryDiskTotalCapacity"))).withColumn("Census_PrimaryDiskTotalCapacity", regexp_replace(col("Census_PrimaryDiskTotalCapacity") , "(\\d{1})(\\d{2})" ,"$1.$2" )).withColumn("Census_PrimaryDiskTotalCapacity", reverse("Census_PrimaryDiskTotalCapacity"))


#Split into train/test
workshop_train, workshop_validate = df_train.randomSplit([0.99, 0.01], seed=1337)

#Create samples of train/test
workshop_train_sample = workshop_train.sample(fraction=0.01, seed=1337)
workshop_validate_sample = workshop_validate.sample(fraction=0.01, seed=1337)

#Temporary save to delta
workshop_train.repartition(5).write.mode("overwrite").format("delta").save(ws_path+"workshop_train_clean_delta")
workshop_train_sample.repartition(5).write.mode("overwrite").format("delta").save(ws_path+"workshop_train_clean_delta_sample")
workshop_validate.repartition(5).write.mode("overwrite").format("delta").save(ws_path+"workshop_validate_clean_delta")
workshop_validate_sample.repartition(5).write.mode("overwrite").format("delta").save(ws_path+"workshop_validate_clean_delta_sample")


#PREP WORKSHOP TRAIN
#Split the training tables into 2 datasets (to join later in the ETL process with the attendees)
basic_first_cols = workshop_train.columns[:34]
basic_last_cols = workshop_train.columns[-5:]
enrich_first_cols = workshop_train.columns[0:1]
enrich_last_cols = workshop_train.columns[34:-5]

train_basic_df = workshop_train.select(basic_first_cols + basic_last_cols)
train_enrich_df = workshop_train.select(enrich_first_cols + enrich_last_cols)

#PREP WORKSHOP TRAIN SAMPLE
#Split the training tables into 2 datasets (to join later in the ETL process with the attendees)
s_basic_first_cols = workshop_train_sample.columns[:34]
s_basic_last_cols = workshop_train_sample.columns[-5:]
s_enrich_first_cols = workshop_train_sample.columns[0:1]
s_enrich_last_cols = workshop_train_sample.columns[34:-5]

s_train_basic_df = workshop_train_sample.select(basic_first_cols + basic_last_cols)
s_train_enrich_df = workshop_train_sample.select(enrich_first_cols + enrich_last_cols)
```

```
#Write the dataframes
train_basic_df.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_train.csv", header=True)
train_enrich_df.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_train_enrich.csv", header=True)

s_train_basic_df.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_train_sample.csv", header=True)
s_train_enrich_df.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_train_enrich_sample.csv", header=True)
s_train_enrich_df.repartition(20).write.mode("overwrite").parquet(ws_path+"workshop_train_enrich_sample_parquet")


workshop_validate.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_validate.csv", header=True)
workshop_validate_sample.repartition(20).write.mode("overwrite").csv(ws_path+"workshop_validate_sample.csv", header=True)
```

```
### Create pre-joined "batch" and "stream"
from pyspark.sql.types import *
from pyspark.sql.functions import *

streamSchema = StructType([
  StructField("MachineIdentifier", StringType(), True),
  StructField("Census_MDC2FormFactor", StringType(), True),
  StructField("Census_DeviceFamily", StringType(), True),
  StructField("Census_OEMNameIdentifier", IntegerType(), True),
  StructField("Census_OEMModelIdentifier", IntegerType(), True),
  StructField("Census_ProcessorCoreCount", IntegerType(), True),
  StructField("Census_ProcessorManufacturerIdentifier", IntegerType(), True),
  StructField("Census_ProcessorModelIdentifier", IntegerType(), True),
  StructField("Census_ProcessorClass", StringType(), True),
  StructField("Census_PrimaryDiskTotalCapacity", StringType(), True),
  StructField("Census_PrimaryDiskTypeName", StringType(), True),
  StructField("Census_SystemVolumeTotalCapacity", IntegerType(), True),
  StructField("Census_HasOpticalDiskDrive", IntegerType(), True),
  StructField("Census_TotalPhysicalRAM", IntegerType(), True),
  StructField("Census_ChassisTypeName", StringType(), True),
  StructField("Census_InternalPrimaryDiagonalDisplaySizeInInches", DoubleType(), True),
  StructField("Census_InternalPrimaryDisplayResolutionHorizontal", IntegerType(), True),
  StructField("Census_InternalPrimaryDisplayResolutionVertical", IntegerType(), True),
  StructField("Census_PowerPlatformRoleName", StringType(), True),
  StructField("Census_InternalBatteryType", StringType(), True),
  StructField("Census_InternalBatteryNumberOfCharges", LongType(), True),
  StructField("Census_OSVersion", StringType(), True),
  StructField("Census_OSArchitecture", StringType(), True),
  StructField("Census_OSBranch", StringType(), True),
  StructField("Census_OSBuildNumber", IntegerType(), True),
  StructField("Census_OSBuildRevision", IntegerType(), True),
  StructField("Census_OSEdition", StringType(), True),
  StructField("Census_OSSkuName", StringType(), True),
  StructField("Census_OSInstallTypeName", StringType(), True),
  StructField("Census_OSInstallLanguageIdentifier", IntegerType(), True),
  StructField("Census_OSUILocaleIdentifier", IntegerType(), True),
  StructField("Census_OSWUAutoUpdateOptionsName", StringType(), True),
  StructField("Census_IsPortableOperatingSystem", IntegerType(), True),
  StructField("Census_GenuineStateName", StringType(), True),
  StructField("Census_ActivationChannel", StringType(), True),
  StructField("Census_IsFlightingInternal", IntegerType(), True),
  StructField("Census_IsFlightsDisabled", IntegerType(), True),
  StructField("Census_FlightRing", StringType(), True),
  StructField("Census_ThresholdOptIn", IntegerType(), True),
  StructField("Census_FirmwareManufacturerIdentifier", IntegerType(), True),
  StructField("Census_FirmwareVersionIdentifier", IntegerType(), True),
  StructField("Census_IsSecureBootEnabled", IntegerType(), True),
  StructField("Census_IsWIMBootEnabled", IntegerType(), True),
  StructField("Census_IsVirtualDevice", IntegerType(), True),
  StructField("Census_IsTouchEnabled", IntegerType(), True),
  StructField("Census_IsPenCapable", IntegerType(), True),
  StructField("Census_IsAlwaysOnAlwaysConnectedCapable", IntegerType(), True)])


###Replicate the join of the _train and _train_enrich tables
enrichedDF = spark.read.schema(streamSchema).csv("/mnt/databricks-workshop-datasets/End-to-End-ML-Lifecycle/workshop_train_enrich.csv")
cleanEnrichedDF = enrichedDF.withColumn("Census_PrimaryDiskTotalCapacity", regexp_replace(col("Census_PrimaryDiskTotalCapacity"), "\.", "")).withColumn("Census_PrimaryDiskTotalCapacity", expr("substring(Census_PrimaryDiskTotalCapacity, 1, length(Census_PrimaryDiskTotalCapacity)-4)"))

defaultBatchDF = spark.read.option("inferSchema", True).option("header", True).csv("/mnt/channelsa-datasets-DONOTEDIT/End-to-End-ML-Lifecycle/workshop_train.csv")
defaultBatchDF = defaultBatchDF.withColumnRenamed("platfrm", "Platform").withColumnRenamed("prcsr", "Processor")
defaultBatchDF = (defaultBatchDF.withColumn("serviceDate", col("serviceDate").cast("string"))
  .withColumn("serviceDateLength", expr("length(serviceDate)"))
  .withColumn("serviceDateFormat", when(col("serviceDateLength") == 6, lit("MMyyyy")).otherwise(lit("Myyyy")))
  .withColumn("serviceDate", expr("to_date(serviceDate, serviceDateFormat)"))
  .drop("serviceDateLength")
  .drop("serviceDateFormat"))
defaultBatchDF = defaultBatchDF.withColumn("recentIncident", from_unixtime(col("recentIncident"), format="yyyy-MM-dd HH:mm:ss"))

fullDF = cleanEnrichedDF.join(defaultBatchDF, on="MachineIdentifier", how="left")

ws_path = "/mnt/channelsa-datasets-DONOTEDIT/End-to-End-ML-Lifecycle/"

fullDF.repartition(50).write.mode("overwrite").partitionBy("serviceDate").parquet(ws_path+"workshop_stream_merged")
fullDF.repartition(50).sample(fraction=0.1, seed=1337).write.mode("overwrite").partitionBy("serviceDate").parquet(ws_path+"workshop_stream_merged_sample")
```
