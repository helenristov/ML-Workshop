# Cleaning Data

In the previous notebook we have seen how we can get data into Spark Dataframes or SQL Tables, and do initial queries and visualizations on it. However, let's dive deeper into our dataset, and create a stable ETL pipeline for our end-users.

As we can see from the sample above, as well as knowledge we have internally, we need to make certain modifications to a few different columns:

* Rename "platfrm" column to "Platform"

* Transforming the "serviceDate" to a Date like "yyyy-MM-01"
* Transform "recentIncident" column from a Unix Timestamp to a "yyyy-MM-dd HH:mm:ss" Timestamp, keeping the column name

### Working with Dates

In this section, we'll get an initial idea on how to work with different ways of processing time information. We'll aim for the following results:
* Transforming the "serviceDate" to a Date like "yyyy-MM-01"
* Transform "recentIncident" column from a Unix Timestamp to a "yyyy-MM-dd HH:mm:ss" Timestamp, keeping the column name

Let's start off with converting "serviceDate". We need to find a transformation in our spark functions that would mold a string into a date format.

Once we have that, are we encountering any issues? If so, can we follow the stack trace and fix it?

```
resolveDF = ...
```

```
### Do not unhide
#############################################################################################
#############################################################################################

from pyspark.sql.functions import *

solvedDF = (solvedDF.withColumn("serviceDate", col("serviceDate").cast("string"))
  .withColumn("serviceDateLength", expr("length(serviceDate)"))
  .withColumn("serviceDateFormat", when(col("serviceDateLength") == 6, lit("MMyyyy")).otherwise(lit("Myyyy")))
  .withColumn("serviceDate", expr("to_date(serviceDate, serviceDateFormat)"))
  .drop("serviceDateLength")
  .drop("serviceDateFormat"))
```

```
display(solvedDF)
```

Now that we have covered off our service date, let's focus on "recentIncident". How do we convert to a particular timestamp format from a unix timestamp?

```
resolveDF = ...
```

```
### Do not unhide
#############################################################################################
#############################################################################################

solvedDF = solvedDF.withColumn("recentIncident", from_unixtime(col("recentIncident"), format="yyyy-MM-dd HH:mm:ss"))
```

```
display(solvedDF)
```

* Free & Open Source.
* Increased query performance over row-based data stores.
* Provides efficient data compression.
* Designed for performance on large data sets.
* Supports limited schema evolution.
* Is a splittable "file format".
* A <a href="https://en.wikipedia.org/wiki/Column-oriented_DBMS" target="_blank">Column-Oriented</a> data store
