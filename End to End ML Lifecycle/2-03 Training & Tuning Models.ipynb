### MLFlow - Assisting with your ML lifecycle

![s](https://databricks.com/wp-content/uploads/2018/06/mlflow.png)

Assuming a very simple model pipeline (which should be pretty bad compare to the one you just built!), let's see how we can apply MLFlow to it as an example.

```
import mlflow
import mlflow.spark
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.classification import DecisionTreeClassifier, LogisticRegression
from pyspark.ml import Pipeline
from pyspark.ml.evaluation import BinaryClassificationEvaluator
from pyspark.mllib.evaluation import BinaryClassificationMetrics

print("MLflow Version:",mlflow.version.VERSION)

def evalMetrics(summary):
  accuracy = summary.accuracy
  auc = summary.areaUnderROC
  return (accuracy, auc)

with mlflow.start_run() as run:
  
  maxIter = 100
  regParam = 0
  elasticNetParam = 0
  
  # Split our dataset into training and testing
  featureColumns = [item[0] for item in df.dtypes if item[1].startswith('int')]
  colsToDrop = ["DefaultBrowsersIdentifier", "Census_IsFlightingInternal", "Census_ThresholdOptIn", "Census_IsWIMBootEnabled", "recentIncident"]
  (train, test) = df.select(*featureColumns).drop(*colsToDrop).na.drop().randomSplit([0.8, 0.2])
  ftCols = train.columns[:-1]
  
  #Also prep the unseen test from before for it to be available for the next notebook.
  unseenCols = ["MachineIdentifier"]+ftCols
  unseenDF.select(*unseenCols).drop(*colsToDrop).na.drop().write.format("delta").mode("overwrite").save("/databricks_workshop/validate_clean")
    
  trainCount = train.count()
  testCount = test.count()
  mlflow.log_param("trainCount", trainCount)
  mlflow.log_param("testCount", testCount)
  
  # First we will use `VectorAssembler` to combine all feature columns into a feature vector (optimized data structure for ML)
  # Then we will scale the values to have a standard mean and deviation
  # Then build a DT model
  # Then build a pipeline with all our stages
  assembler = VectorAssembler(inputCols=ftCols, outputCol="featureVector")
  scaler = StandardScaler(inputCol="featureVector", outputCol="features", withStd=True, withMean=False)
  lr = LogisticRegression(featuresCol="features", labelCol="HasDetections", maxIter=maxIter, regParam=regParam, elasticNetParam=elasticNetParam)
  pipeline = Pipeline(stages=[assembler, scaler, lr])

  # Fit the model
  model = pipeline.fit(train)
  lrModel = model.stages[-1]

  #Get metrics
  (accuracy, auc) = evalMetrics(lrModel.summary)

  # Log mlflow parameters to the LR model
  mlflow.log_param("maxIter", maxIter)
  mlflow.log_param("regParam", regParam)
  mlflow.log_metric("elasticNetParam", elasticNetParam)

  # Log mlflow metrics for the LR model
  mlflow.log_metric("accuracy", accuracy)
  mlflow.log_metric("auc", auc)

  # Log model generated
  mlflow.spark.log_model(model, "lrModel")
```

Alternatively to the above, you can also build an scikit-learn model!

```
import mlflow
import mlflow.sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, accuracy_score

def evalMetrics(actual, pred):
  accuracy = accuracy_score(actual, pred)
  auc = roc_auc_score(actual, pred)
  return (accuracy, auc)

with mlflow.start_run() as run:

  # Split our dataset into training and testing
  featureColumns = [item[0] for item in df.dtypes if item[1].startswith('int')]
  colsToDrop = ["DefaultBrowsersIdentifier", "Census_IsFlightingInternal", "Census_ThresholdOptIn", "Census_IsWIMBootEnabled", "recentIncident"]
  
  #print(df.head())
  
  (train, test) = df.select(*featureColumns).drop(*colsToDrop).na.drop().limit(100000).randomSplit([0.8, 0.2])
  ftCols = train.columns[:-1]

  #Transform it to Pandas
  train = train.toPandas()
  test = test.toPandas()

  X_train = train.iloc[:, :-1]
  y_train = train.iloc[:, -1:]
  X_test = test.iloc[:, :-1]
  y_test = test.iloc[:, -1:]
  trainCount = train.shape[0]
  testCount = test.shape[0]
  mlflow.log_param("trainCount", trainCount)
  mlflow.log_param("testCount", testCount)
  
  #Define parameters
  maxIter = 105
  regParam = 1
  penalty = "l2"
  
  # Log mlflow parameters to the LR model
  mlflow.log_param("maxIter", maxIter)
  mlflow.log_param("regParam", regParam)
  mlflow.log_param("elasticNetParam", penalty)
   #Train the model
  clf = LogisticRegression(C=regParam, max_iter=maxIter, penalty="l2")
  clf.fit(X_train, y_train)
  
  #Predict train and test
  y_train_pred = clf.predict(X_train)
  y_test_pred = clf.predict(X_test)
  
  acc_train, auc_train = evalMetrics(y_train, y_train_pred)
  acc_test, auc_test = evalMetrics(y_test, y_test_pred)
  
  # Log mlflow metrics for the LR model
  mlflow.log_metric("accuracy", acc_train)
  mlflow.log_metric("auc", auc_train)
  
  mlflow.sklearn.log_model(clf, "lrModelSK")
```
