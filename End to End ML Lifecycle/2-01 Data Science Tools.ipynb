### Data Science Tools

As you have seen in the ETL ecosystem, there are a variety of tools to handle different parts of the pipeline. For Data Science, you have a similar story. You have different companies tackling different parts of this ecosystem, in order to help organizations build value on top of their existing data.

Using Gartner as a standard benchmark again, we can have a look at their most recent Data Science and Machine Learning Platforms Magic Quadrant.

![s](https://www.kdnuggets.com/images/gartner-mq-data-science-ml-platforms-2019-640.jpg)

Depending on your use case, you might want to use one or a combination of different tools. If you want to do visual data preparation and analytics, users might want to use a desktop-based tool such as Alteryx, RapidMiner or KNIME, use a traditional software like SAS, or do visual pushdown through Dataiku. If they are exclusively focused on building data projects using Python, maybe they want to use Domino in the cloud, or manage their environments using the Anaconda stack. If you're focused on auto machine learning, maybe you focus on AzureML, DataRobot, or H20. The cloud providers like Microsoft and Google also provide an extensive stack on building data projects within their cloud.

However, there are still a large variety of data scientists who don't necessarily use any tools for their model building. They rely purely on open source technologies, installing and running these either on their local machine, a server setup on-premise in their organizations, or on a shared VM in the cloud. As such, they might be working with data extracts in CSV files, which they read from a local or shared drive in python, where they do data processing and modeling on potentially small samples due to CPU or Memory limitations, save a model back on the shared drive, and manually re-run the same scripts on a weekly or monthly basis to produce scored CSV datasets.

In Databricks, a Data Scientist has the ability to easily connect to data sources using native connectors, quickly and easily munge data at scale using optimized, auto-scaling Spark clusters in a PaaS environment, build (distributed) machine learning models, and productionize their workloads, all in one single Unified Analytics Platform.

Furthermore, the Azure ecosystem contains a variety of services specialized for companies wishing to put Machine Learning and AI into practice. From a variety of pre-trained models which can be applied easily to any dataset, to support of all the popular open-source and properietary frameworks available for customization of data science pipelines. Whether these are built through specialized Machine Learning VMs, or Azure Databricks and Azure Machine Learning for enterprise applications, users have the ability to leverage powerful compute infrastructure adapted to their use cases. These projects can be flexibly deployed, in an iterative process, either to cloud or on-prem solutions, or to edge devices, to quickly reach the positive business outcomes that were the main drivers of the use cases.
